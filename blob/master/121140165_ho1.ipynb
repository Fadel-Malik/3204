B. Pengaruh Pergeseran Fase Terhadap Posisi Gelombang

Pergeseran fase pada suatu sinyal sinus atau kosinus akan menyebabkan seluruh gelombang bergeser ke kiri atau ke kanan sepanjang sumbu waktu. Ini seperti menggeser seluruh grafik ke arah sumbu x.

Pergeseran Fase Positif: Jika kita menambahkan nilai positif pada argumen fungsi sinus atau kosinus (yaitu bagian dalam kurung), maka gelombang akan bergeser ke kiri. Artinya, puncak dan lembah gelombang akan mencapai nilai maksimum atau minimum lebih cepat dibandingkan dengan gelombang yang tidak bergeser.
Pergeseran Fase Negatif: Sebaliknya, jika kita menambahkan nilai negatif pada argumen fungsi, maka gelombang akan bergeser ke kanan. Puncak dan lembah gelombang akan terlambat mencapai nilai ekstremnya.

C. Amplitudo adalah ukuran seberapa jauh suatu gelombang menyimpang dari posisi kesetimbangannya. Sederhananya, amplitudo menentukan "tinggi" atau "rendah" nya suatu gelombang.

Pengaruh Amplitudo:

Amplitudo Besar: Semakin besar amplitudo suatu gelombang, semakin jauh pula gelombang tersebut menyimpang dari garis tengah (sumbu x). Ini berarti gelombang akan terlihat lebih "tinggi" dan "rendah".

Amplitudo Kecil: Sebaliknya, jika amplitudo kecil, maka gelombang akan lebih "datar" atau "rendah". Variasi nilai gelombang akan lebih kecil.

D. Pergeseran fase adalah pergeseran horizontal dari sebuah gelombang sinus atau kosinus sepanjang sumbu waktu. Ini seperti menggeser seluruh grafik ke kiri atau ke kanan.

Pengaruh Pergeseran Fase:

Pergeseran ke Kiri: Jika kita menambahkan nilai positif pada argumen fungsi sinus atau kosinus, maka gelombang akan bergeser ke kiri. Artinya, puncak dan lembah gelombang akan mencapai nilai maksimum atau minimum lebih cepat dibandingkan dengan gelombang yang tidak bergeser.
Pergeseran ke Kanan: Sebaliknya, jika kita menambahkan nilai negatif pada argumen fungsi, maka gelombang akan bergeser ke kanan. Puncak dan lembah gelombang akan terlambat mencapai nilai ekstremnya.

```python
#No. 4
import numpy as np
import matplotlib.pyplot as plt

# Parameter sinyal
fs = 1000  # Frekuensi sampling
t = np.arange(0, 2, 1/fs)  # Waktu

# Fungsi untuk membuat sinyal sinus
def generate_sinusoid(A, f, phi):
    return A * np.sin(2 * np.pi * f * t + phi)

# Buat sinyal y1, y2, dan y3
A1, f1, phi1 = 2, 3, 0
A2, f2, phi2 = 1, 4, np.pi/4
y1 = generate_sinusoid(A1, f1, phi1)
y2 = generate_sinusoid(A2, f2, phi2)

# Kombinasi sinyal
y5 = y1 + y2

# Plot sinyal
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, y1, label='y1')
plt.title('Sinyal Individu')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, y2, label='y2')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, y5, label='y5 = y1 + y2')
plt.title('Sinyal Hasil Penjumlahan')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Apa yang terjadi ketika menjumlahkan dua gelombang sinus?

Interferensi: Hasil penjumlahan dua gelombang sinus disebut interferensi. Interferensi dapat bersifat konstruktif atau destruktif.

    Interferensi Konstruktif: Jika kedua gelombang berada pada fase yang sama (puncak bertemu puncak, lembah bertemu lembah) saat dijumlahkan, amplitudo gelombang resultan akan meningkat.

    Interferensi Destruktif: Jika kedua gelombang berada pada fase yang berlawanan (puncak bertemu lembah) saat dijumlahkan, amplitudo gelombang resultan akan berkurang, bahkan bisa menjadi nol pada beberapa titik.

Bentuk Gelombang Baru: Bentuk gelombang hasil penjumlahan akan tergantung pada:
    Amplitudo: Amplitudo gelombang resultan akan dipengaruhi oleh amplitudo kedua gelombang awal dan fase relatif antara keduanya.
    Frekuensi: Jika frekuensi kedua gelombang sama, gelombang resultan akan memiliki frekuensi yang sama. Jika berbeda, gelombang resultan akan memiliki komponen frekuensi dari kedua gelombang tersebut.
    
    Fase: Perbedaan fase antara kedua gelombang akan menentukan bentuk gelombang resultan. Jika fase berbeda 180 derajat, maka akan terjadi interferensi destruktif total.

```python
#No. 5
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import resample

# Parameter sinyal (sesuaikan dengan NIM Anda)
durasi = 165  # Sesuaikan dengan 3 digit terakhir NIM
sampling_rate = 150
noise_level = 0.65  # Sesuaikan dengan 2 digit terakhir NIM
heart_rate = 80
random_state = 240925

# Simulasi sinyal EKG
np.random.seed(random_state)  # Set seed untuk reprodusibilitas
t = np.arange(0, durasi, 1/sampling_rate)  # Vektor waktu
ecg_signal = np.sin(2 * np.pi * heart_rate/60 * t) + noise_level * np.random.randn(len(t))

# Fungsi downsampling
def downsample(signal, factor):
    return signal[::factor]

# Downsampling dengan berbagai faktor
factors = [2, 3, 5, 10, 20]
signals = [downsample(ecg_signal, factor) for factor in factors]

# Plot sinyal
plt.figure(figsize=(12, 8))
plt.plot(t, ecg_signal, label='Sinyal Asli')
for i, factor in enumerate(factors):
    plt.plot(t[::factor], signals[i], label=f'Downsampled x{factor}')
plt.xlabel('Waktu (s)')
plt.ylabel('Amplitudo')
plt.title('Efek Downsampling pada Sinyal EKG')
plt.legend()
plt.grid(True)
plt.show()
```

Parameter 'order' dalam konteks filtering, biasanya merujuk pada tingkat kompleksitas atau orde dari filter. Ini menunjukkan jumlah koefisien yang digunakan dalam filter untuk melakukan operasi filtering.

Apa yang dimaksud dengan 'order'? Semakin tinggi nilai order, semakin kompleks filter tersebut. Filter dengan order yang lebih tinggi umumnya dapat memberikan atenuasi yang lebih baik pada frekuensi yang tidak diinginkan, namun juga dapat memperkenalkan lebih banyak perhitungan dan latensi.

Apa yang terjadi ketika mengubah nilai 'order'?
Order rendah: Filter dengan order rendah cenderung memiliki respon frekuensi yang kurang tajam dan selektif. Mereka mungkin tidak dapat menghilangkan noise atau distorsi frekuensi tinggi dengan efektif.

Order tinggi: Filter dengan order tinggi memiliki respon frekuensi yang lebih tajam dan selektif. Mereka dapat menghilangkan noise dan distorsi frekuensi tinggi dengan lebih baik, namun juga dapat memperkenalkan efek samping seperti ringing atau overshoot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter, freqz

# Parameter
fs = 1000  # Frekuensi sampling
cutoff = 200  # Frekuensi cutoff
order = [2, 4, 8]  # Berbagai nilai order

# Buat sinyal
t = np.arange(0, 1, 1/fs)
x = np.sin(50 * 2 * np.pi * t) + 0.5 * np.random.randn(len(t))

# Fungsi untuk desain filter Butterworth low-pass
def butter_lowpass(cutoff, fs, order):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

# Terapkan filter dan plot hasil
plt.figure(figsize=(12, 8))
for i, n in enumerate(order):
    b, a = butter_lowpass(cutoff, fs, n)
    y = lfilter(b, a, x)

    # Plot sinyal hasil filter
    plt.subplot(3, 1, i+1)
    plt.plot(t, y, label=f'Ordo {n}')
    plt.legend()
    plt.grid(True)

    # Plot respons frekuensi (opsional)
    w, h = freqz(b, a, worN=2000)
    plt.figure()
    plt.plot(w, 20 * np.log10(abs(h)))
    plt.xlabel('Frekuensi (rad/sample)')
    plt.ylabel('Amplitudo (dB)')
    plt.title(f"Respon frekuensi untuk ordo {n}")
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#No. 7
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Parameter
fs = 100  # Frekuensi sampling (Hz)
t = np.arange(0, 60, 1/fs)  # Waktu dalam detik
f_resp = 0.2  # Frekuensi respirasi (Hz)
noise_level = 0.2  # Tingkat noise

# Generate sinyal respirasi dengan noise
x = np.sin(2 * np.pi * f_resp * t) + noise_level * np.random.randn(len(t))

# Desain filter band-pass
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='bandpass')
    return b, a

# Tentukan frekuensi cutoff
lowcut = 0.1  # Hz
highcut = 0.3  # Hz

# Terapkan filter
b, a = butter_bandpass(lowcut, highcut, fs, order=5)
y = lfilter(b, a, x)

# Plot hasil
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(t, x, label='Sinyal Asli')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, y, label='Sinyal Hasil Filter')
plt.legend()
plt.xlabel('Waktu (detik)')
plt.ylabel('Amplitudo')
plt.title('Filter Band-Pass untuk Sinyal Respirasi')
plt.show()
```
